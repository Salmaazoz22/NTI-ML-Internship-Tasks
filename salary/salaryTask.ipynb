# Part3

```python
#  Convert Salary to Numeric
df['Salary'] = df['Salary'].replace('[\$,]', '', regex=True).astype(float)
df['Salary'].head(10)
```

```python
#  Normalize Education and Gender
df['Education Level'] = df['Education Level'].astype(str).str.title().str.strip()
df['Gender'] = df['Gender'].str.title().str.strip()
df[['Education Level', 'Gender']].head(10)

```

```python
# Fill Missing Gender with Mode
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Gender'].isnull().sum()
```

```python
#  Fill Missing Experience with Median
df['Years of Experience'] = df['Years of Experience'].fillna(df['Years of Experience'].median())
df['Years of Experience'].isnull().sum()
```

```python
# Remove Extreme Salaries > 500,000
df = df[df['Salary'] <= 500000]
df['Salary'].max()
```

```python
#  Drop Columns with >90% Missing
threshold = len(df) * 0.9
df = df.dropna(thresh=threshold, axis=1)
```

```python
#  Print Shape and Summary After Cleaning
df.shape, df.info()
```

# **Part6**

```python
# IsSenior Feature
df['IsSenior'] = df['Job Title'].str.contains("Senior|Sr\.|Lead|Principal", case=False, na=False).astype(int)
df[['Job Title', 'IsSenior']].head(10)
```

```python
#  IsManager Feature
df['IsManager'] = df['Job Title'].str.contains("Manager", case=False, na=False).astype(int)
df[['Job Title', 'IsManager']].head(10)
```

```python
#  Experience Level Binning
def exp_level(x):
    if x < 2: return 'Entry'
    elif x < 5: return 'Junior'
    elif x < 10: return 'Mid'
    elif x < 20: return 'Senior'
    else: return 'Veteran'
df['ExpLevel'] = df['Years of Experience'].apply(exp_level)
df[['Years of Experience', 'ExpLevel']].head(10)
```

```python
#  Z-score Normalization
df['Salary_zscore'] = (df['Salary'] - df['Salary'].mean()) / df['Salary'].std()
df[['Salary', 'Salary_zscore']].head(10)
```

```python
#  Gender Binary Column
df['GenderBin'] = df['Gender'].map({'Male': 0, 'Female': 1})
df[['Gender', 'GenderBin']].head(10)
```

```python
# JobType Column
def job_type(title):
    title = title.lower()
    if 'engineer' in title: return 'Engineer'
    elif 'scientist' in title: return 'Scientist'
    elif 'analyst' in title: return 'Analyst'
    else: return 'Other'
df['JobType'] = df['Job Title'].apply(job_type)
df[['Job Title', 'JobType']].head(10)
```

```python
# Salary per Year of Experience
df['SalaryPerYear'] = df['Salary'] / df['Years of Experience'].replace(0, np.nan)
df[['Salary', 'Years of Experience', 'SalaryPerYear']].head(10)

```

```python
#  One-hot Encode country
country_dummies = pd.get_dummies(df['Country'], prefix='Country')
df = pd.concat([df, country_dummies], axis=1)
# Show the original Country column with its new one-hot encoded columns
df[['Country'] + [col for col in df.columns if col.startswith('Country_')]].head(10)


```

```python
#  IsTech Binary
df['IsTech'] = df['Job Title'].str.contains("Engineer|Developer|Tech|Software|Data", case=False, na=False).astype(int)
df[['Job Title', 'IsTech']].head(10)

```

```python
#  Normalize Years of Experience
df['YearsOfExp_Normalized'] = (df['Years of Experience'] - df['Years of Experience'].min()) / (df['Years of Experience'].max() - df['Years of Experience'].min())
df[['Years of Experience', 'YearsOfExp_Normalized']].head(10)
```
